```python
f = open("../response (3).json")
hoaxy_td = json.load(f)
hoaxy_td
```

```python
sites_list = []
sites_dict = {}
for item in hoaxy_td['articles']:
    if "//www" in item['canonical_url']:
        el = item['canonical_url'].split(".")[1]
    else:
        el = item['canonical_url'].split("//")[1].split(".")[0]
    sites_list.append(el)
    if el not in sites_dict:
        sites_dict[el] = 1
    else:
        sites_dict[el] += 1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(sites_dict.keys(), sites_dict.values())

plt.show()
```

```python
sites_dict
```

```python
double_dict = {}
trial_sites = []
for i,it in enumerate(sites_list):
    if it not in double_dict:
        double_dict[it] = 1
    elif double_dict[it] <=2500:
        double_dict[it] += 1
        trial_sites.append(hoaxy_td['articles'][i]['canonical_url'])
        
len(trial_sites)
```

```python
import re

fake_list = []
legit_list = []
for i in range(514):
    with open("../../Dropbox/Fake News Project/DataSets/dataset/fake/{}_fake.txt".format(i), encoding="utf8") as f:
        lines = re.sub(r'[\W]', r' ', f.read())
        if len(lines) != 0:
            fake_list.append(lines)
for i in range(756):
    with open("../../Dropbox/Fake News Project/DataSets/dataset/legit/{}_legit.txt".format(i), encoding="utf8") as f:
        lines = f.read()
        if len(lines) != 0:
            legit_list.append(lines)
```
